---- Preprocessing ----

```python
train_dt = train_dt.drop(columns = ['id', 'CustomerId', 'Surname'],axis = 1)
test_dt = test_dt.drop(columns = ['id', 'CustomerId', 'Surname'],axis = 1)
```

```python
X_train_encoder = pd.get_dummies(train_dt, columns=['Geography',
                                       'Gender',
                                       'IsActiveMember',
                                       'HasCrCard',
                                       'NumOfProducts'])
X_train_encoder
```

```python
move_last = X_train_encoder.pop('Exited')
X_train_encoder['Exited'] = move_last

X_train_encoder
```

```python
X_test_encoder = pd.get_dummies(test_dt, columns=['Geography',
                                       'Gender',
                                       'IsActiveMember',
                                       'HasCrCard',
                                       'NumOfProducts'])

X_test_encoder
```

```python
plt.figure(figsize=(23,17))
mask = np.triu(np.ones_like(X_train_encoder.corr()))
round_ = round(X_train_encoder.corr(),2)
sns.heatmap(round_, mask=mask, annot=True, cmap='Blues')
```

```python
X = X_train_encoder.drop(['Exited'], axis=1)
y = X_train_encoder['Exited']
```

```python
y
```

---- Building Model ----

```python
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV, cross_val_score
from sklearn.metrics import ConfusionMatrixDisplay,confusion_matrix,classification_report, roc_auc_score
from sklearn.ensemble import VotingClassifier

from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
```

```python
# To maintain the equal proportion of the both classes we use stratify. stratify parameter will preserve the proportion of target as in original dataset, in the train and test datasets as well.
X_train,X_test,y_train,y_test = train_test_split(X, y, random_state=42, stratify=y)
```

XGBoost

```python
xgb_model = XGBClassifier(objective='binary:logistic',seed=42)
xgb_model.fit(X_train,
            y_train,
            verbose=True,
            early_stopping_rounds=10,
            eval_metric='auc',
            eval_set=[(X_test, y_test)])
```

```python
xgb_pred = xgb_model.predict(X_test)
old_xgb_roc_auc = roc_auc_score(y_test,xgb_pred)
print('ROC-AUC before Hyperparamerters tuning',old_xgb_roc_auc)
```

```python
print('-'*15,'XGBoost before GridSearchCV','-'*15)

cm = confusion_matrix(y_test, xgb_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Not Left', 'Left'])
disp.plot()
plt.show()
print(classification_report(y_test, xgb_pred))
```

```python
# XGBoost hyperparameters tuning
xgb_params = {
     'max_depth': [3, 4, 5],
     'learning_rate': [0.05, 0.01, 0.1],
     'gamma': [0, 0.25, 1.0],
     'reg_lambda': [0, 1.0, 5, 10],
     'scale_pos_weight': [1, 3, 5],
     'subsample' : [0.5,1]
     }

xgb_opt_params = GridSearchCV(
     estimator=XGBClassifier(objective='binary:logistic',
                                 colsample_bytree=0.5),
     param_grid= xgb_params,
     scoring='roc_auc',
     verbose=False,
     cv=3
 )

xgb_opt_params.fit(X_train,
                        y_train,
                        early_stopping_rounds=10,
                        eval_metric='auc',
                        eval_set=[(X_test, y_test)],
                        verbose=False)

print(xgb_opt_params.best_params_)
```

```python
# XGBoost with best parameters
best_xgb = XGBClassifier(
                         objective='binary:logistic',
                         colsample_bytree=0.5,
                         gamma=0,
                         learning_rate=0.1,
                         max_depth=5,
                         reg_lambda=10.0,
                         scale_pos_weight=3,
                         subsample=1
                         )

best_xgb.fit(X_train,
               y_train,
               verbose=False,
               early_stopping_rounds=10,
               eval_metric='auc',
               eval_set=[(X_test,y_test)]
               )

last_result = best_xgb.best_score
print("Last result:", last_result)

best_xgb_pred = best_xgb.predict(X_test)
xgb_roc_auc = roc_auc_score(y_test,best_xgb.predict_proba(X_test)[:,1])
```

```python
print('-'*15,'XGBoost after GridSearchCV','-'*15)
cm = confusion_matrix(y_test, best_xgb_pred, labels=best_xgb.classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Not Left', 'Left'])
disp.plot()
plt.show()
print(classification_report(y_test, best_xgb_pred))
```

LightGBM

```python
# Hyperparameter tuning

lgbm_params = {
    'learning_rate': [0.05, 0.01, .1],
    'max_depth' :  [4, 5, 6],
    'subsample' : [0.5, 0.75, 1],
    'min_child_weight' : [0.1, 1, 5, 15],
    'reg_lambda' : [0.1, 5, 10, 20],
    'reg_alpha' : [0.1, 5, 10]
    }

lgbm_opt_params = GridSearchCV(
    estimator=LGBMClassifier(boosting_type='gbdt', objective='binary',verbosity=-1),
    param_grid = lgbm_params,
    scoring='roc_auc',
    verbose=False ,
    cv=3
    )

lgbm_opt_params.fit(X_train,
                        y_train,
                        eval_metric='auc',
                        eval_set=[(X_test, y_test)]
                        )

print(lgbm_opt_params.best_params_)
```

```python
# LightGBM with best parameters
best_lgbm = LGBMClassifier(
                               boosting_type='gbdt',
                               objective='binary',
                               learning_rate=0.1,
                               max_depth=4,
                               min_child_weight=1,
                               reg_lambda=0.1,
                               reg_alpha=0.1,
                               subsample=0.5,
                               verbose=-1
                               )

best_lgbm.fit(X_train,
               y_train,
               eval_metric='auc',
               eval_set=[(X_test,y_test)])

last_result = best_lgbm.best_score_
print("Last result:", last_result)

best_lgbm_pred = best_lgbm.predict(X_test)
lgbm_roc_auc = roc_auc_score(y_test,best_lgbm.predict_proba(X_test)[:,1])
```

```python
cm = confusion_matrix(y_test, best_lgbm_pred, labels=best_lgbm.classes_)
print('-'*15,'LightGBM after GridSearchCV','-'*15)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Not Left', 'Left'])
disp.plot()
plt.show()
print(classification_report(y_test, best_lgbm_pred))
```

Catboost

```python
# Catboost hyperparameters tuning
cat_params = {'iterations': [500],
          'depth': [3, 4, 5,],
          'loss_function': ['Logloss', 'CrossEntropy'],
          'l2_leaf_reg': np.logspace(-20, -19, 3),
          'leaf_estimation_iterations': [10]
         }

cat_opt_params = GridSearchCV(
     estimator=CatBoostClassifier(),
     param_grid = cat_params,
     scoring='roc_auc',
     cv=3
 )

cat_opt_params.fit(X_train,
                        y_train,
                        eval_set=[(X_test, y_test)],
                        verbose=False)

print(cat_opt_params.best_params_)
```

```python
# Catboost best parameters
best_cat = CatBoostClassifier(
                              depth=5,
                              iterations= 500,
                              l2_leaf_reg=1e-20,
                              leaf_estimation_iterations=10,
                              loss_function= 'Logloss',
                              eval_metric='AUC',
                              verbose = False
                              )

best_cat.fit(X_train,
               y_train,
               eval_set=[(X_test,y_test)])

last_result = best_cat.get_best_score()
print("Last result:", last_result)

best_cat_pred = best_cat.predict(X_test)
cat_roc_auc = roc_auc_score(y_test,best_cat.predict_proba(X_test)[:,1])
```

```python
cm = confusion_matrix(y_test, best_cat_pred, labels=best_cat.classes_)
print('-'*15,'Catboost after GridSearchCV','-'*15)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Not Left', 'Left'])
disp.plot()
plt.show()
print(classification_report(y_test, best_cat_pred))
```

```python
print('XGBoost ROC-AUC probability :',xgb_roc_auc)
print('LightGBM ROC-AUC probability :',lgbm_roc_auc)
print('Catboost ROC-AUC probability :',cat_roc_auc)
```

Voting

```python
voting_ensemble = VotingClassifier(estimators=[('XGBoost',best_xgb),('LightGBM',best_lgbm),('CatBoost',best_cat)], voting='soft')

voting_ensemble.fit(X_train, y_train)

vote_y_pred = voting_ensemble.predict(X_test)
```

```python
#Voting Probability ROC-AUC
vote_roc_auc = roc_auc_score(y_test,voting_ensemble.predict_proba(X_test)[:,1])
print(vote_roc_auc)
```

```python
cm = confusion_matrix(y_test, vote_y_pred, labels=voting_ensemble.classes_)
print('-'*20,'Voting Ensemble','-'*20)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Not Left', 'Left'])
disp.plot()
plt.show()
print(classification_report(y_test, vote_y_pred))
```
